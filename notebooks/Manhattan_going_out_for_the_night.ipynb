```python
rideshares = pd.read_parquet(r'C:\Users\EvanSpiller\Documents\DS BootCamp\Capstone\data\HVFHV_Compressed\Combined\full_compressed_HVFHV_2022_2023.parquet')
rideshares2022Q4night = rideshares.loc[
    (
       ((rideshares['Month'] == 10) |(rideshares['Month'] == 11) | (rideshares['Month'] == 12)) &
    ((rideshares['Year'] == 2022) | (rideshares['Year'] == 2022))
     & (rideshares['Hour']>= 20)
    )
]
rideshares2022Q4night['Date']= pd.to_datetime(rideshares2022Q4night[['Year', 'Month', 'Day']])
rideshares2022Q4night['Day_of_week'] = rideshares2022Q4night['Date'].dt.strftime('%A')
rideshares2022Q4night = rideshares2022Q4night.loc[
    (
        (rideshares2022Q4night['Day_of_week'] == 'Friday') | 
        (rideshares2022Q4night['Day_of_week'] == 'Saturday') 
    )
]
```

```python
ridesharesq420192022nightsatsun = pd.concat([rideshares2022Q4night, rideshares2019Q4night], axis=0)
```

```python
zones = pd.read_csv(r'C:\Users\EvanSpiller\Documents\DS BootCamp\Capstone\data\taxi+_zone_lookup.csv')
```

```python
ridesharesq420192022nightsatsun = pd.merge(ridesharesq420192022nightsatsun, zones, left_on='DOLocationID', right_on='LocationID', how='inner')
```

```python
ridesharesq420192022nightsatsun = ridesharesq420192022nightsatsun.loc[ridesharesq420192022nightsatsun['Borough']=='Manhattan',:]
```

```python
ridesharesq420192022nightsatsun
```

```python
groupby = ridesharesq420192022nightsatsun.groupby(['PULocationID','Year'])['trips'].sum()
```

```python
groupby = groupby.unstack()
```

```python
groupby = groupby.reset_index()
```

```python
groupby['YoY'] = (groupby[2022] - groupby[2019])/groupby[2019]
```

```python
groupby['YoY'] = groupby['YoY']*100
```
